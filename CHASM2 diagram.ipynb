## CHASM2 diagram

This notebook contains miscellaneous figures showing how CHASM2 works (Figure 1). 

### Cross-validation pie chart

Create a diagram showing how cross-validation is done.

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
flatui_pal = sns.color_palette(flatui)
plt.pie([.1]*10, colors=[flatui_pal[3]]+[flatui_pal[2]]*9, explode=[.25]+[0]*9, labels=['fold {0}'.format(i) for i in range(1,11)])
plt.legend(loc='best')
plt.axis('equal')
plt.savefig('figures/cross_validation_pie_chart.pdf')
```

### Score distribution

Plot one minus the cumulative distribution function based on the simulated mutations.

```python
null_df = pd.read_table('data/chasm2_null_distribution.txt').dropna(subset=['CHASM2_genome'])
with sns.axes_style('ticks'), sns.plotting_context('paper', font_scale=1.5):
    plt.plot(null_df.score, null_df['CHASM2_genome'])
    plt.ylabel('1-CDF')
    plt.xlabel('gwCHASM2')
    sns.despine()
    plt.xlim((0, 1))
    plt.gca().set_aspect('equal')
    plt.savefig('figures/null_distribution.pdf')
```
